```python
from SCreator import FileReader
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
eggs = []
dir_path = 'data2k17/Healthy_dark/'
reader = FileReader.FileReader(dir_path)
```

```python
eggs = reader.read_files()
```

```python
for i in range(len(eggs)):
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
    plt.xlabel('Wavelength [nm]')
    plt.ylabel('Reflectance')
    for j in range(5):
        data = eggs[i].samples[j].data
        plt.scatter(data["wavelength"],data["reflectance"], s=1)
    fig.savefig('imgs/scatter_graphs/Healthy_dark/graph_' + eggs[i].samples[0].part_class + '.png' , dpi=500)
    plt.clf()
```

```python
#biale chore
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (% / 100)')
for i in range(5):
    data = eggs[10].samples[i].data
    plt.scatter(data["wavelength"],data["reflectance"], s=1)
plt.grid(True)
plt.xticks(np.arange(min(eggs[10].samples[1].data["wavelength"]), max(eggs[10].samples[1].data["wavelength"])+1, 20.0))
fig.savefig('imgs/scatter_graphs/ill_white_ticked_graph.png' , dpi=700)
plt.show()
```
